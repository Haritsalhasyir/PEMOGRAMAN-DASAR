```python
# Menampilkan bilangan ganjil 1 sampai 200 dengan while
i = 1
while i <= 200:
    print(i)
    i += 2
```

```python
#input angka untuk menghitung perkalian menggunakan kondisi while
angka = int(input("tabel perkalian:"))
x = 1
while x <= 10:
  print(angka, "x", x, "=", angka * x)
  x += 1
```
